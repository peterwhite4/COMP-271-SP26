```python
def m():
    print(" M       M")
    print(" M M   M M")
    print(" M  M M  M")
    print(" M   M   M")
    print(" M       M")
    print(" M       M")
    print()


def i():
    print("     I    ")
    print("     I    ")
    print("     I    ")
    print("     I    ")
    print("     I    ")
    print("     I    ")
    print()


def s():
    print("   SSSSSSS")
    print(" SS       ")
    print(" SS       ")
    print("  SSSSSS  ")
    print("        SS")
    print("        SS")
    print(" SSSSSSS  ")
    print()


def p():
    print(" PPPPPPPP ")
    print(" P      PP")
    print(" P     PP ")
    print(" PPPPPP   ")
    print(" P        ")
    print(" P        ")
    print(" P        ")
    print()


def ssi():
    s()
    s()
    i()


def ssissi():
    ssi()
    ssi()


def pp():
    p()
    p()


def mississippi():
    m()
    i()
    ssissi()
    pp()
    i()


def main():
    mississippi()


main()
```
